### Analysing the data  

```python
# Qn.1: How many are there in each Team and the percentage splitting with respect to the total employees.
```

```python
#Counting
count = df["Team"].value_counts()
print("No of employees in each team:\n",count)
```

```python
#Visualizing
plt.figure(figsize=(60,30),dpi=100)
x=count.index
y=count.values
plt.bar(x, y, color='green',width=0.5)
plt.xticks(fontsize=35)
plt.xticks(rotation=90)
plt.yticks(fontsize=35)
plt.title('Number of employees in each team',size=65)
plt.xlabel('Team',size=45)
plt.ylabel('Number of employees',size=45)
plt.show()
```

```python
#Total number of teams.
print("Total number of teams =",len(count))
```

```python
#Percentage split of employees by each team.
percentage_split = count/ len(df) * 100
print("The percentage split of employees with respect to the total employees:\n",percentage_split)
```

```python
#Visualizing
plt.figure(figsize=(60,30),dpi=100)
x=percentage_split.index
y=percentage_split.values
plt.bar(x, y, color=['green','red'],width=0.5)
plt.xticks(fontsize=35)
plt.xticks(rotation=90)
plt.yticks(fontsize=35)
plt.title('Percentage split of employees in each team',size=65)
plt.xlabel('Team',size=45)
plt.ylabel('Percentage',size=45)
plt.show()
```

```python
# Qn.2: Segregate the employees w.r.t different positions.
```

```python
#position grouping
Position_data = df.groupby(['Player position','Player name']).sum()
Position_data
```

```python
# Count of employees working in each postions,
position_count=df['Player position'].value_counts()
print("Number of players in different positions are:")
position_count
```

```python
#visualizing the results.
plt.bar(position_count.index,position_count.values,color='green')
plt.xlabel("Player position")
plt.ylabel("Number of employees/players")
plt.title("Employee Position Distribution")
```

```python
# Qn.3: Find from which age group most of the employees belong to.
```

```python
#grouping
age_group = df["Age"].value_counts()
print("Different age group of players are:")
age_group
```

```python
#which age group most of the employees
age_max=age_group.head(1).index
print("Most of the employees belong to age group of",age_max[0])
```

```python
#visualizing
plt.figure(figsize=(22,22))
plt.pie(age_group,labels=age_group.index,autopct='%0.2f%%',radius=1)
plt.title("Distribution of players by age group",size=35)
plt.show()
```

```python
# Qn.4: Find out under which team and position, spending in terms of salary is high.
```

```python
team_salary=df.groupby(['Team','Player position'])['Salary in $'].sum()
team_salary
```

```python
Highest_salary = team_salary.idxmax()
print("Highest salary is in:",Highest_salary)
print("The {} team and {} player position is spending high in terms of salary!".format (Highest_salary[0],Highest_salary[1]))
```

```python
# Qn.5: Find if there is any correlation between age and salary , represent it visually.
```

```python
#Finding correlation
correlation=df['Age'].corr(df["Salary in $"])
print("The correlation between age and salary is",correlation)

#Scatter plot visualization
age=df['Age']
salary=df['Salary in $']
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.title("Age v/s Salary")
sns.scatterplot(x=age,y=salary,hue=age,palette='viridis')
plt.show()
```

#### From the above scatter plot even though we can observe no linear relation ship between age and salary of players but it is evident from the plot that as age increases some players are having high salary may be due to their experience in the field.
